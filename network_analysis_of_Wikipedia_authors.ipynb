### Network Analysis of Wikipedia Authors

Wiki_data.txt contains a random selection of 1000 Wikipedia authors, with a directed link from author i to author j if author i voted on author j (see snap.stanford.edu/data/wiki-Vote.html for more details). Remove the empty nodes that have no edges from/to.

1. In a directed graph, how many vertices and edges are there?


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import nxviz as nv
from collections import Counter

df = pd.read_csv('Wiki_data.txt')
print(df.head())
G = nx.DiGraph()
G.add_nodes_from(df.iloc[:,0])
G.add_nodes_from(df.iloc[:,1])
print('number of nodes = ', G.number_of_nodes())

G = nx.from_pandas_edgelist(df, 'FromNodeId', 'ToNodeId', edge_attr=True, create_using=nx.DiGraph())
print('number of edges = ', G.number_of_edges())
```

2. Plot the directed graph for the first 50 vertices according to their indices in the as- cending order. The vertex labels do not need to be the same as the indices.

```python
first = df.loc[df["FromNodeId"] <= 50]
data = first.loc[df["ToNodeId"] <= 50]
print(data.head())

H = nx.DiGraph()
H.add_nodes_from(data.iloc[:,0])
H.add_nodes_from(data.iloc[:,1])
H = nx.from_pandas_edgelist(data, 'FromNodeId', 'ToNodeId', edge_attr=True, create_using=nx.DiGraph())
nx.draw(H, with_labels=True)
plt.show()
```

9. Using the undirected graph with all vertices plot a histogram of the separation (the length of the shortest path between two vertices) between all vertices based on separations from 1 to 20. That is the bin n(s) in the histogram contains the number of pairs with sep- aration s (i.e., no need to normalize the histogram). Write the number of unique pairs with separations from 1 to 20.

Here maximum separation length is 7.

```python
b = nx.all_pairs_shortest_path_length(U)

#here b is a generator object so i convert it to a DataFrame
c = pd.DataFrame(b)

# unique pairs
print(c.iloc[:,1].head())

# here I'm calculating the number of occurrences 
# for the length of the shortest path between two vertices
list1 =[]
for i in range(542):
    z = Counter(c[1][i].values())
    list1.append(z.values())
    
list1 = pd.DataFrame(list1)
sum = list1.sum()
sum = np.array(sum/2) 
# divided over 2 because in the adjacency_matrix,
# for example, there is an edge- author 8 is connected to author 20
# and there's also another edge- author 20 is connected to author 8
# both of these edges are actually the same edge in an undirected graph

print('separation length:    [1, \t 2, \t  3, \t   4, \t    5, \t     6,      7]')
print('number of occurences:', list(sum[1:]))
plt.figure(figsize=(6,6))
plt.bar([1, 2, 3, 4, 5, 6, 7], sum[1:])
plt.xlim((0.5,7.5))
plt.title("Histogram of the separation")
plt.xlabel("separation \n (the length of the shortest path between two vertices)")
plt.ylabel("number of occurrences")
```
